```python
def image_padding(img_whole):
    img = np.zeros((1024,1024))
    img = img.astype(np.float32)
    
#     img_whole = cv2.resize(img_whole, (1024,img_whole.shape[0]))
    h,w = img_whole.shape
    print(img_whole.shape)

    if h>1024:
        img = img_whole
        pass
    elif (1024 - h) != 0:
        gap = int((1024 - h)/2)
        img[gap:gap+h,:] = img_whole
    elif (1024 - w) != 0:
        gap = int((1024 - w)/2)
        img[:,gap:gap+w] = img_whole
   

    return img
```
